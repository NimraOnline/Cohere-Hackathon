# Removing Arabic and Chunking

```python
import pandas as pd
df = pd.read_csv('/content/tafsir.csv')
df = df.drop('Unnamed: 0', axis=1)
```

**Removing Arabic**

```python
import pandas as pd
import re

# Function to remove Arabic characters using regular expression
def remove_arabic(text):
    # Remove Arabic characters
    arabic_pattern = re.compile('[\u0600-\u06FF\s]+')
    text_without_arabic = arabic_pattern.sub(' ', text)

    # Remove numeric sequences
    numeric_pattern = re.compile(r'\b\d+\.\s*')
    text_without_numbers = numeric_pattern.sub(' ', text_without_arabic)

    # Remove consecutive dots
    dot_pattern = re.compile(r'\. \s*')
    text_without_dots = dot_pattern.sub(' ', text_without_numbers)

    pattern = re.compile(r'\.\((.*?)\)')
    result = re.sub(pattern, r'"\1."', text_without_dots)

    return result

# Apply the function to the 'Text' column
df['Tafsir'] = df['Tafsir'].apply(remove_arabic)

# Display the result
df.head()
```

```python
# Find the largest tafsir size in terms of words
largest_tafsir_words = df['Tafsir'].apply(lambda x: len(str(x).split())).max()

print(f"The largest tafsir size is: {largest_tafsir_words} words.")
```

**Chunking**

```python
import pandas as pd
import re

def split_into_chunks(text, target_chunk_size=500):
    words = text.split()
    chunks = []
    current_chunk = ""

    for word in words:
        if len(current_chunk.split()) + len(word.split()) <= target_chunk_size:
            current_chunk += word + " "
        else:
            chunks.append(current_chunk.strip())
            current_chunk = word + " "

    if current_chunk:
        chunks.append(current_chunk.strip())

    return chunks

# Apply the function to the 'Tafsir' column
df['Tafsir'] = df['Tafsir'].apply(lambda x: split_into_chunks(str(x)))

# Explode the DataFrame to handle multiple paragraphs
df_exploded = df.explode('Tafsir')

# Reset index
df_exploded = df_exploded.reset_index(drop=True)

# Display the result
df_exploded
```
